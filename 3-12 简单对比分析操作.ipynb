````python
import pandas as pd

import numpy as np

df = pd.read_csv('data/HR.csv')

print(df.head(5))
````

````python
df = df.dropna(axis=0,how="any") 
````

````python
df
````

### last_evaluation异常值处理

````python
df = df[df['last_evaluation'] <=1]
````

````python
df[df["salary"] !="nme"][df['department']!="sale"]
````

上面这种括号选择最多嵌套两层，嵌套两层以上会报出警告

```
Boolean Series key will be reindexed to match DataFrame index

```

### 以部门为单位进行一次对比分析

groupby后面跟一个聚合函数，当同一个部门的聚合到一起之后的处理方式

````python
df.groupby('department').mean()
````

### 拉出其中的几列来进行分析

index全取，属性列选取

````python
df.loc[:,["last_evaluation","department"]].groupby("department").mean()
````

### 自定义函数进行对比

自定义一个极差函数: 最大值-最小值

````python
df.loc[:,["average_monthly_hours","department"]].groupby("department")["average_monthly_hours"].apply(lambda x:x.max()-x.min())
````
